```python
def clean_text(text):
    """
    Define a Cleaning Function:


    Cleans text by removing HTML tags, dummy data, and special characters.

    Args:
        text (str): The input text.

    Returns:
        str: The cleaned text.
    """
    # Check if text is a valid string or bytestring
    if isinstance(text, (str, bytes)):
        # 1. Remove HTML tags using Beautiful Soup
        soup = BeautifulSoup(text, 'html.parser')
        text = soup.get_text()

        # 2. Remove dummy data or specific patterns using regex
        # Add dummy data patterns with variations including spaces
        dummy_patterns = [
            r"s*<s*ul\s*>s*", r"s*<s*/ul\s*>s*", r"s*<s*li\s*>s*", r"s*<s*/li\s*>s*",
            r"s*<s*p\s*>s*\s*:\s*s*<s*/p\s*>s*", r"s*<s*br\s*/?>s*",
            r"s*<s*span\s*>s*", r"s*<s*/span\s*>s*", r"<div\s*>", r"</div\s*>",
            r"<strong\s*>", r"< /strong\s*>", r"<strong>", r"</strong>",
            r"< /li\s*>",
            r"<ol\s*>", r"< /ol\s*>",
            r"< /div\s*>",
            r"<em>", r"</em>",
            r"< ul\s*>", r"< /ul\s*>", r"< li\s*>", r"< /li\s*>",
            r"< div\s*>", r"< /div\s*>",
            r"s*<s*p\s*>s*benefits\s*:\s*s*<s*/p\s*>s*",
            r"s*<s*span\s*>s*<s*p\s*>s*<s*/p\s*>s*<s*/span\s*>s*",  # New pattern for "<span> <p> </p>"
            r"s*<s*p\s*>s*<s*strong\s*>s*<s*span\s+style\s*=\s*['\"]{1,3}[^'\"]*text-decorat[^'\"]*['\"]{1,3}\s*>s*",
            r"<s*span\s*>s*<s*p\s*>s*’s*<s*p\s*>s*<s*strong\s*>s*<s*/p\s*>"  # Pattern for "< span > < p > ’ < p > < strong > < /p >"
        ]  # Pattern for "<p> <strong> <span style=..."  # New pattern for "<span> <p> </p>"
        #\s*: This is added to match zero or more spaces before or after the tags



        for pattern in dummy_patterns:
                text = re.sub(pattern, "", text)

        # 3. Remove special characters and punctuation (optional)
        text = re.sub(r'[^\w\s]<>', '', text)

        # 4. Convert to lowercase
        text = text.lower()

        # 5. Tokenize the text
        tokens = word_tokenize(text)

        # 6. Remove stop words
        stop_words = set(stopwords.words('english'))
        filtered_tokens = [w for w in tokens if not w in stop_words]

        # 7. Join the tokens back into a string
        cleaned_text = ' '.join(filtered_tokens)

        return cleaned_text
    # Return original value (e.g., NaN) if not a string or bytestring
    return text
```

```python
def calculate_annual_salary(row):
    """
    Calculates annual salary based on salary range and period.
    """
    salary_range = row['salary_range']
    salary_period = row['salary_period']

    # Extract minimum and maximum salary from salary_range
    try:
        min_salary, max_salary = map(int, salary_range.replace(',', '').split('-'))
    except ValueError:
        return np.nan  # Handle cases where salary_range is not in the expected format

    # Calculate annual salary based on salary period
    if salary_period == 'Per Year':
        annual_salary = (min_salary + max_salary) / 2  # Average of the range
    elif salary_period == 'Per Month':
        annual_salary = (min_salary + max_salary) / 2 * 12
    elif salary_period == 'Per Week':
        annual_salary = (min_salary + max_salary) / 2 * 52
    elif salary_period == 'Per Day':
        annual_salary = (min_salary + max_salary) / 2 * 365 # Consider a year as 365 days
    else:
        annual_salary = np.nan  # Handle unknown salary periods

    return annual_salary

# Apply the function to create the new column 'annual_salary'
df_egypt['annual_salary'] = df_egypt.apply(calculate_annual_salary, axis=1)
```

```python


def extract_skills(text):
    # turn to lowercase, split on common delimiters
    parts = re.split(r'[,\.;:/\-\n•|]+', str(text).lower())
    # strip whitespace, drop very short tokens or the word "and"/"or"
    skills = [p.strip() for p in parts
              if len(p.strip())>2 and p.strip() not in {"and","or","must","skills"}]
    return skills

# create the job_skills column (a list of skills for each posting)
df_egypt['job_skills'] = df_egypt['job_requirements'].apply(extract_skills)

# if you want the skill‑counts **by** job_title, explode + groupby:
exploded = df_egypt.explode('job_skills').dropna(subset=['job_skills'])
skill_counts_by_title = (
    exploded
      .groupby(['job_title','job_skills'])
      .size()
      .reset_index(name='frequency')
      .sort_values(['job_title','frequency'], ascending=[True,False])
)

```

```python
import pandas as pd

# نأخذ فقط الأعمدة المهمة ونشيل الـ NaN
titles_skills = df_egypt[['job_title', 'job_skills']].dropna()

# دالة لتنظيف وتحليل المهارات
def clean_skills(text):
    return (
        pd.Series(text)
        .astype(str)  # تحويل كل القيم إلى سترينج
        .str.replace(r"<.*?>", "", regex=True)  # إزالة التاجات
        .str.replace(r"[\r\n•·]", ",", regex=True)  # استبدال الفواصل المختلفة بـ ,
        .str.split(",")  # تقسيم المهارات
        .explode()  # فردها لسطر لكل مهارة
        .str.strip()  # إزالة المسافات
        .str.title()  # توحيد الشكل
    )

# جدول لتجميع النتائج
results = []

# نلف على كل عنوان وظيفة فريد
for title in titles_skills['job_title'].dropna().unique():
    # نفلتر الصفوف اللي فيها نفس العنوان
    temp = titles_skills[titles_skills['job_title'] == title]

    # ننضف ونفصل المهارات
    all_skills = clean_skills(temp['job_skills']).dropna()
    all_skills = all_skills[all_skills != ""]

    # نجيب أكتر 4 مهارات مكررة
    top_skills = all_skills.value_counts().head(4).index.tolist()

    # نضيفهم للنتيجة
    results.append({
        'job_title': title,
        'top_skills': ", ".join(top_skills)
    })

# تحويل النتائج إلى DataFrame
skills_by_title_df = pd.DataFrame(results)

# لو حابب تدمج النتيجة مع df_egypt:
df_egypt = df_egypt.merge(skills_by_title_df, on='job_title', how='left')


```

```python
import pandas as pd
import numpy as np
import re

df= pd.read_excel(r"C:\Users\Abdelrahman Yasser\Desktop\DEPI SA\Mega Project\Wuzzuf Dataset\Wuzzuf_semi_cleaned.xlsx")
# Industry table
df['industry'] = df['industry'].replace('Select', np.nan)


# salary range table

# Assuming your DataFrame is called df
# Step 1: Identify rows where salary_range doesn't contain '-'
mask = ~df['salary_range'].astype(str).str.contains('-')

# Step 2: Set those rows to "Unknown" and clear salary fields
df.loc[mask, 'salary_range'] = 'Unknown'
df.loc[mask, ['Min_salary', 'Max_salary', 'annual_salary']] = np.nan

print(f"Marked {mask.sum()} rows as 'Unknown'")

df



# Define the list of junk patterns
junk_patterns = [
    r'\(\s*\)',              # empty parentheses
    r'-+',                   # multiple dashes
    r'\.+',                  # multiple dots
    r'\d{1,2}-\d{1,2}',      # number ranges like 10-12
    r'\d+',                  # standalone numbers
    r'â€“', r'æœ¬.*?ç§‘',     # corrupted characters
    r'[^\w\s\']',            # remove weird symbols but keep words, spaces, and apostrophes
]

# Combine them into one regex
junk_regex = re.compile('|'.join(junk_patterns))

def clean_requirements(text):
    if not isinstance(text, str):
        return np.nan

    # Remove junk
    cleaned = junk_regex.sub(' ', text)

    # Remove extra spaces
    cleaned = re.sub(r'\s+', ' ', cleaned).strip()

    # If nothing meaningful remains, return NaN
    return cleaned if cleaned and re.search(r'[a-zA-Z]', cleaned) else np.nan

# Apply to your column
df['job_requirements'] = df['job_requirements'].apply(clean_requirements)

df['description'] = df['description'].apply(clean_requirements)

columns_to_clean = ['description', 'job_requirements','top_skills','1st_skill', '2nd_skill', '3rd_skill', '4th_skill']  # add more if needed

df[columns_to_clean] = df[columns_to_clean].apply(lambda col: col.apply(clean_requirements))

df.info()

df=df.reset_index(drop=True)

df.to_excel(r"C:\Users\Abdelrahman Yasser\Desktop\DEPI SA\Mega Project\cleaned_Wuzzuf1.xlsx", index=False)

import os
print(os.getcwd())


df= pd.read_excel(r"C:\Users\Abdelrahman Yasser\Desktop\DEPI SA\Mega Project\cleaned_Wuzzuf.xlsx")
df.info()

df = df.fillna({col: 'unknown' if df[col].dtype == 'object' else 0 for col in df.columns})
```
